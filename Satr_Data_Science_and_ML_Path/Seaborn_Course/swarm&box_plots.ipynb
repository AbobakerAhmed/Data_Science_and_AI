```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the data

```python
data = pd.read_csv("mpg_raw.csv")
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(data,x="origin", y="mpg", hue= "cylinders");
```
