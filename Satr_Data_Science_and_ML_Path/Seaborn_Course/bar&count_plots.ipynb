## **Bar & Count Plots**

```python
data = pd.read_csv("mpg_raw.csv")
```

```python
sns.barplot(data, x= "origin", y= "horsepower");
```

### CountPlot

```python
sns.countplot(data, x= "origin");
```

What is the relation between model year and cylinder?
Creating a new column which will contains the mpg level

```python
data["mpg_level"] = data["mpg"].apply(lambda x: "High" if x>30 else "Low" if x<=15 else "Median")
```

```python
plt.figure(figsize = (12,5))
sns.countplot(data, x = "model_year", hue= "mpg_level");
```

we note that high mpg cars are made increasingly with the time, and the low is stoped before 78, and the mediam is the most but it gonna to to be stoped


That is means the economic cards is going to grow with time, and the high and median gas consuming cars are going to be less and less with the time.
