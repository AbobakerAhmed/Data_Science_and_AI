### 1- Importing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2- Read the data using load_dataset

```python
df = sns.load_dataset("tips")
```

---
### Take a look at the data

```python
df.info() # Columns and their contents
```

```python
df.describe(include="all") # quick statistics about each column
```

```python
print(f"{df.head()}\n\n{df.tail()}") # see some of the data rows
```

---
### 4- Make a Histogram for Tips

```python
plt.figure(figsize= (12,5))
plt.title("Tips Distribution Over The Week")
sns.set_style("whitegrid")
sns.histplot(
    data = df,
    x= "tip",
    hue= "day",
    multiple = "stack",
    element= "step",
    kde= True
);
```

We note that:
- Thu has the heighest frequencies
- Tips are focuced from 1 to 6

---
### 4- Make Displot for Days

```python
from turtle import title


sns.displot(
    data = df,
    x= "day",
    hue= "sex",
    hue_order= ["Female", "Male"],
    multiple = "dodge",
    shrink= 0.9,
    palette= "Set1",
);
```

we note that:
- females in Thursday are more than males
- Other ways, males are more than females

---
### 5- DCDF Chart

```python
sns.set_style("darkgrid")
sns.set(font_scale= 1.25)

sns.ecdfplot(
    data =df,
    x= "tip",
    hue= "time",
    palette= "summer",
    linewidth= 3
);
```
