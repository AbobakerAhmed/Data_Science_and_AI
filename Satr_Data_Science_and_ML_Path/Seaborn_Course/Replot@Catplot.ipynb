# **Replot & Catplot**

### Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the data

```python
data = pd.read_csv("mpg_raw.csv")
```

```python
sns.relplot(data = data, x="model_year", y ="mpg")
```

```python
sns.relplot(data, x="horsepower", y="mpg", hue="origin", size ="weight");
```

```python
sns.catplot(data, x="origin", y="horsepower", kind="bar", hue = "cylinders", alpha = 0.7, height = 4, palette= "bright");
```
