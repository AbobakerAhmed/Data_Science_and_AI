# **JointPlot & PairPlot**

### Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the data

```python
data = pd.read_csv("mpg_raw.csv")
```

#### Creating The JointPlot

```python
sns.jointplot(data = data, x = "mpg", y = "weight", kind = "scatter");
```

```python
sns.pairplot(data, hue = "origin", kind = "reg");
```

```python
plt.bar(data.model_year, data.mpg);
```
