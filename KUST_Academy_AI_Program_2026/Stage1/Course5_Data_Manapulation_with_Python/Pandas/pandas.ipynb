    - average of all old values in columnName1 without ignoring null vlues (if there is one the result will be NaN)

    - two new columns of column3Name, one with mean and the second with the std of all old values in column3Name related to a single value of the selected column/s in groupby



##### Tranformation
Transformation is different from aggregation. Where agg() returns a single value per column, so one row per group, tranform() returns an object that is the same size as the group. Essentially, it broadcasts the function you supply over the grouped dataframe, returning a new dataframe. This makes combining data later easy.

Transformation (transform): Non-destructive. It keeps the original number of rows, pasting the calculated value next to every original record.

df.groupby(..anything..).transform(np.stMethod) or df["columnName"].transform(np.method): stMethods = [average, mean, nanmean, ...]
this will apply that stMethod to the selected column and return a new data frame contains each value of that selected column beside the calculated value of all old vlaues of a single group value

##### Filltering
The GroupBy object has build in support for filtering groups as well. It's often that you'll want to group by some feature, then make some transformation to the groups, then drop certain groups as part of your cleaning routines. The filter() function takes in a function which it applies to each group dataframe and returns either a True or a False, depending upon whether that group should be included in the results.

df.groupby('columnName').filter(function): return the value calculated in the fucntion with filltering the new data frame with a spcific conditions in the function
