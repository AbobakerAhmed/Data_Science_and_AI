##### Anonymous Functions - "lambda" Key Word

```python
# old definition of a function
def double(parameter):
    returned_value = parameter*2
    return returned_value

result_of_long_style = double(2) # x=4

# define a function using lambda
double_using_lambda = lambda parameter:  parameter*2 # = returned_value
result_from_lambda = double_using_lambda(2)

print(f"are results the same? {result_of_long_style == result_from_lambda}") # they are the same

# in the old style, we was define the funtion then call it, now we can define it and call it directily
result = (lambda x: x*2)(2)
print(f"x * 2 = {result}")
```

##### "reverce" and "key" Optional Parameter

```python
print("_" * 30 + "\"reverse\" key word"+"_" *80)
# reverse: sort in decending order
# reverse in sort

list1.sort(reverse= True)
print(f"list 1 in decending order: {list1}")

# reverse in sorted
list3 = sorted(list2, reverse=True)
print(f"list 2 without change: {list2}")
print(f"list 3, which is list 2 in decending order: {list3}")
print()

print("_" * 30 + "\"key\" key word"+"_" *80)
# key, which alow a function then do this steps:
# 1- assign a funcion to the variable
# 2- sort there result after appling that fuction
# 3- inverse the fucion and return each value as it in its new place

def absolute(x):
    if x>=0:
        return x
    return -x

# sort abslute values of the list
print(f"list 1 before: {list1}")
list1.sort(key = absolute)
print(f"list 1 after sorting its abslute values: {list1}")

list3 = sorted(list2, key= absolute)
print(f"original list 2 is : {list2}")
print(f"list 3 after sorting list 2 absolute values: {list3}")


print("_" * 30 + "use \"key\", \"lambda\", and \"reverse\"key words"+"_" *80)
list1.sort(reverse= True, key= lambda x: abs(x))
print(f"list 1 in decending absolute order: {list1}")
list3 = sorted(list2, reverse= True, key=  lambda x: abs(x))
print(f"list 3 after sorting list2 in aboslute decending order: {list3}")
```

##### Sort a Dectionary

```python
L = ['E', 'F', 'B', 'A', 'D', 'I', 'I', 'C', 'B', 'A', 'D', 'D', 'E', 'D']

d = {}
for x in L:
    if x in d:
        d[x] += 1
    else:
        d[x] = 1


# y = sorted(d.keys()) # sort keys in alphabit order
# y = sorted(d) # also will sort be keys
# y= sorted(d, key= lambda x: d[x]) # sort values in acending order

y = sorted(d, key=lambda k: d[k], reverse=True) # sort be the values in decending order (highest first)
for k in y:
    print("{} appears {} times".format(k, d[k]))
```

##### Breaking Ties: Second Sorting

```python
# assume you have a list of values you have to sort,
# but some of them have the same value of the key variable of sorting.
# which one should be the first? ok, we can sort them by another key variable
# that exactly what called breaking ties
# we pass the key variables in a tuple and returned it in a lambda statment

lst = [
    (2,3,10),
    (1,5,3),
    (2,2,5),
    (1,2,6)
]
print(f"After sorting with the first element, then second element\n{sorted(lst)}") # that is the default value
print(f"After sorting with the second element, then first element\n{sorted(lst, key= lambda tup: (tup[1], tup[0]) )}")  # returned value of lambda (second element, first element)
print(f"After sorting with the sum of tuple, then second element\n{sorted(lst, key= lambda tup: (sum(tup), tup[1]) )}")  # returned value of lambda (sum, second element)
print(f"After sorting with the sum of tuple, then second element in Decending Order using reverse\n{sorted(lst, key= lambda tup: (sum(tup), tup[1]), reverse= True )}")  # returned value of lambda (sum, second element)
print(f"After sorting with the sum of tuple, then second element in Decending Order using (-)\n{sorted(lst, key= lambda tup: (-sum(tup), -tup[1]) )}")  # same as above
```

##### When to use a Lambda Expression

**Use lambda If**
- When the logic fits on a single line
- When accessing nested data
- When sorting by multiple criteria (Tie-breaking)
- When applying simple arithmetic tricks (Numerical only)

**Don't Use lambda If**
- Complex Logic
- When Handling Complex Reverse Sorting

#### 1.

To start, define a function called strip_punctuation which takes one parameter, a string which represents a word, and removes characters considered punctuation from everywhere in the word. (Hint: remember the .replace() method for strings.)

```python
punctuation_chars = ["'", '"', ",", ".", "!", ":", ";", '#', '@']

def strip_punctuation(word):
    for char in word:
        if char in punctuation_chars:
            word = word.replace(char, "")
    return word
assert strip_punctuation("'Hello!") == "Hello" # test
```

#### 2.

Next, define a function called get_pos which takes one parameter, a string which represents one or more sentences, and calculates how many words in the string are considered positive words. Use the list, positive_words to determine what words will count as positive. The function should return a positive integer - how many occurrences there are of positive words in the text. Note that all of the words in positive_words are lower cased, so you’ll need to convert all the words in the input string to lower case as well.

```python
# list of positive words to use
positive_words = []
with open("final_assignment_files/positive_words.txt", "r") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            positive_words.append(lin.strip())

def get_pos(sentences):
    pos_count = 0
    for word in sentences.split(" "):
        if strip_punctuation(word.lower()) in positive_words:
            pos_count += 1
    return pos_count

test = "That was good to see you again. I did like your clothes, they were nice on you. Great day I hope to you."
assert(get_pos(test)) == 4 # good, like, nice, Great
```

```python
# list of negative words to use
negative_words = []
with open("final_assignment_files/negative_words.txt") as pos_f:
    for lin in pos_f:
        if lin[0] != ';' and lin[0] != '\n':
            negative_words.append(lin.strip())

def get_neg(sentences):
    neg_count = 0
    for word in sentences.split(" "):
        if strip_punctuation(word.lower()) in negative_words:
            neg_count += 1
    return neg_count
test = "That was good to see you again. I did like your clothes, they were nice on you. Great day I hope to you."
assert(get_neg(test)) == 0
```

```python
# order: (the text of a tweet, the number of retweets of that tweet, and the number of replies to that tweet)
twts_file = open("final_assignment_files/project_twitter_data.csv", "r")

twts_dec = {} # here we will extract the content of the file

header = twts_file.readline().strip().split(',') # take the header out
# header = ["tweet_text", "retweet_count", "reply_count"]
assert header == ["tweet_text", "retweet_count", "reply_count"]

# now set header elements as keys of the dec
for item in header: 
    twts_dec[item] = [] # this list later will holds the data of each twt in their order

# now extract twts data and assign their values to the dectionary
for twt_data in twts_file.readlines():  # twt data here will contains txt, retweets, replies

    # now let's extract the twt_data it self
    feilds = twt_data.strip().split(',') # take the line and extract the content in a list [twt text, retweets, replies]

    # now assign each value to the list of its field in the same order
    for title in header:
        twts_dec[title].append(feilds[ header.index(title) ])
        # header.index(item) sould return the index of the item in the keys of the dectionary,
        # which represents the order in the list of the twt data =  [twt text, retweets, replys]

twts_file.close() #now we don't need this file any more, all of that data in the twt_dec


# opon the results file and write the header
results_file = open("final_assignment_files/resulting_data.csv", "w") # it will delete 
# we should write in the order:
# Number of Retweets, Number of Replies, Positive Score, Negative Score, Net Score
results_file.write("Index,Number of Retweets,Number of Replies,Positive Score,Negative Score,Net Score\n")
# now lest calculate the p, n, words in the text
twt_index = 0
for twt in twts_dec['tweet_text']:
    pos_words = get_pos(twt) # geting positive words
    neg_words = get_neg(twt) # get negative words
    retweets = twts_dec['retweet_count'][ twt_index ] # save no. retweets
    replays = twts_dec['reply_count'][ twt_index ] # save no. replays
    # now we have all what we need (retweets, replies, recount of pos, count of neg, and nets)

    # write then in the result file in that order
    results_file.write(f"{retweets},{replays},{pos_words},{neg_words},{pos_words - neg_words}\n")
    twt_index += 1

results_file.close()
```
