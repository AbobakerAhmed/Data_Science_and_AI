### Module 2: Classes

The state (atributes) of an object represents those things that the object knows about itself. The state is stored in instance variables 

#### Class Definition

```python
# class heding
class ClassName: # class naming should be UpperCase (each word in the name start with a capital letter)
    """Class docstring:
        Describe what is this this class
        (This is also the way docstrings work in functions.)
    """


    def __init__(self): # constructor
        """
        This initializer method,
        often referred to as the constructor,
        is automatically called whenever a new instance of ClassName is created.
        It gives the programmer the opportunity to set up the attributes required within the new instance
        by giving them their initial state values.
        """

        """
        The self parameter (you could choose any other name, but nobody ever does!)
        is automatically set to reference the newly created object that needs to be initialized.
        """

        self.state1 = "Here we define the atributes of the instance, this is the first one" 
        self.state2 = "this is the second atribute of this class"
```

```python
# use the class
instance1 = ClassName() # automaticly call __init__ method
instance2 = ClassName()
instance3 = instance1
print(f"I'm state1 of instance1: {instance1.state1}")
print(f"I'm state2 of instance2: {instance1.state1}")
print(f"Are instance1 and instance2 refer to the same memory address? {instance1 is instance2}")
print(f"Are instance1 and instance3 refer to the same memory address? {instance1 is instance3}")
```

####  Adding Other Methods to a Class

All methods defined in a class that operate on objects of that class will have self as their first parameter. Again, this serves as a reference to the object itself which in turn gives access to the state data inside the object.

```python
class ClassName:
    """ docstring """

    def __init__(self, init_state1, init_state2):
        """ constructor"""
        self.state1 = init_state1
        self.state2 = init_state2

    # define new functions of the class
    def add_states(self):
        return self.state1 + self.state2
    

    def get_state1(self):
        return self.state1

    def get_state2(self):
        return self.state2
```

```python
# use it
inst1 = ClassName("val_of_state1","val_of_state2")
print(f"I'm state1 using methods: {inst1.get_state1()}")
print(f"I'm state2 using methods: {inst1.get_state2()}")
print(f"I'm state1 + state2 using methods: {inst1.add_states()}")
```

#### Converting an Object to a String (__ str __)

```python
class ClassName:
    """ docstring"""

    def __init__(self, init_state1, init_state2):
        """ constructor"""
        self.state1 = init_state1
        self.state2 = init_state2

    def __str__(self):
        return "state1= {}, state2= {}".format(self.state1, self.state2)

# use it:
inist = ClassName("state1 of inist1", "state2 of inist")
print(f"use __str__ method: {inist.__str__()}")
```

#### Sort List of Instances

```python
# using key
class TestSorting:
    def __init__(self, name , x):
        self.x = x
        self.name = name

    def get_x(self):
        return self.x

    def get_name(self):
        return self.name



instance1 = TestSorting("ins1",1)
instance2 = TestSorting("ins2",2)
instance3 = TestSorting("ins3",3)

lst = [instance1,instance2,instance3]
lst.sort(key= TestSorting.get_x ) # pass the method by the class
print(f"sorted instance accending: {[ins.get_name() for ins in lst]}")
lst.sort(key= lambda ins: ins.get_x() , reverse =True) # using lambda, sort decending
print(f"sorted instance accending: {[ins.get_name() for ins in lst]}")
```

```python
# using __lt__
class TestSorting:
    def __init__(self, name , x,y):
        self.x = x
        self.y = y
        self.name = name

    def get_x(self):
        return self.x

    def get_name(self):
        return self.name
    
    # lest than method, used for sorting
    def __lt__(self, others): # self refere to that instance and other for other instances
        return self.x < others.x

    # we can also use greater than method to do the reverse
    def __gt__(self, others):
        return self.y > others.y
    
    # if both exist, the default to use is __lt__
    # important note: if you put any print statment in any of them it will not be called by default as a key of sorthing


instance1 = TestSorting("ins1",1,3)
instance2 = TestSorting("ins2",2,2)
instance3 = TestSorting("ins3",3,1)
lst = [instance1,instance2,instance3]

lst.sort() # without use key,
print(f"sort the list accending without key: {[ins.get_name() for ins in lst]}") 
print(f"Is instance1.x < instance2.x? {instance1 < instance2}") # evaluate instance1.__lt__(instance2)
```

#### Public and Private Instance Variables

This idea is called abstraction or (incapsolation or something like that)

```python
class PP: # Publec and Private

    def __init__(self, var1, var2, var3):
        self.var1 = var1 # public
        self._var2 = var2 # it is just a sign that varibale should be private but it is public :)
        self.__var3 = var3 # it is private, but it can be accessed some how, we will see
    
    def get_var1(self):
        return self.var1
    
    def get_var2(self):
        return self._var2

    def get_var3(self):
        return self.__var3
    

# use them 
instance = PP("val of var1", "val of var2", "val of var3")

print(f"var 1 using its method: {instance.get_var1()}")
print(f"var 1 using its name: {instance.var1}")
print()

print(f"var 2 using its method: {instance.get_var2()}")
print(f"var 2 using its name: {instance._var2}")
print()

print(f"var 3 using its method: {instance.get_var3()}")
#print(f"var 3 using its name: {instance.__var3}") # AttributeError: 'PP' object has no attribute '__var3'
print(f"How to access private varibales like var3? {instance._PP__var3}") # int the form intanceName._ClassName__stateName
```
