#### Sort List of Instances

```python
# using key
class TestSorting:
    def __init__(self, name , x):
        self.x = x
        self.name = name

    def get_x(self):
        return self.x

    def get_name(self):
        return self.name



instance1 = TestSorting("ins1",1)
instance2 = TestSorting("ins2",2)
instance3 = TestSorting("ins3",3)

lst = [instance1,instance2,instance3]
lst.sort(key= TestSorting.get_x ) # pass the method by the class
print(f"sorted instance accending: {[ins.get_name() for ins in lst]}")
lst.sort(key= lambda ins: ins.get_x() , reverse =True) # using lambda, sort decending
print(f"sorted instance accending: {[ins.get_name() for ins in lst]}")
```

#### Public and Private Instance Variables

This idea is called abstraction or (incapsolation or something like that)

```python
class PP: # Publec and Private

    def __init__(self, var1, var2, var3):
        self.var1 = var1 # public
        self._var2 = var2 # it is just a sign that varibale should be private but it is public :)
        self.__var3 = var3 # it is private, but it can be accessed some how, we will see
    
    def get_var1(self):
        return self.var1
    
    def get_var2(self):
        return self._var2

    def get_var3(self):
        return self.__var3
    

# use them 
instance = PP("val of var1", "val of var2", "val of var3")

print(f"var 1 using its method: {instance.get_var1()}")
print(f"var 1 using its name: {instance.var1}")
print()

print(f"var 2 using its method: {instance.get_var2()}")
print(f"var 2 using its name: {instance._var2}")
print()

print(f"var 3 using its method: {instance.get_var3()}")
#print(f"var 3 using its name: {instance.__var3}") # AttributeError: 'PP' object has no attribute '__var3'
print(f"How to access private varibales like var3? {instance._PP__var3}") # int the form intanceName._ClassName__stateName
```

#### Inheriting Variables and Methods

```python
class FatherClass:
    father_class_var1 = "I'm var1 in father class"

    def __init__(self, state1):
        self.state1 = state1
    
    def print_from_father(self):
        return ("Hello I'm method form the father class")


# son calss
class SonClass(FatherClass): # like this SoneClass extends FatherClass

    def __init__(self, state2): # in case son dosen't have __init__ method, the father's one will be invoked as a contructor automatically
        self.state2 = state2

    def method_form_son(self):
        print("Hello I'm a method form the son class")
    


# use them
father = FatherClass(5)
son = SonClass(10)

print(f"state1 of father: {father.state1}")
print(f"Use father method: {father.print_from_father()}")
print()
print(f"state2 of son: {son.state2}")
print(f"son method: {son.method_form_son()}")
print()

# use father variables and methods
print(f"father class var from son: {son.father_class_var1}")
print(f"father method using son: {son.print_from_father()}")
```

#### Overriding Methods & Invoking the Parent Class’s Method

```python
class Parent:
    def __init__(self, color):
        self.color = color

    def change_color(self):
        self.color = "white" # in the perent we will change the color to white
    

class Son(Parent):
    def __init__(self, color):
        super().__init__(color) # calling the parent method when this method is invoked
                                # that is called Invoking the Parent Class’s


    # override change_color method
    def change_color(self):
        self.color = "Black"


# use 
son = Son("Blue")
print(f"Son Color: {son.color}")
son.change_color() # it will call the method in the son class not parent
print(f"After changing the color: {son.color}")
son.change_color()
```

#### Function Wrapping and Decorators

 - We cannot directly manipulate the behavior of a function, we can wrap it in another function that does something before or after the original function is called or change the arguments that a function takes. This is called function wrapping.
 - This kind of function wrapping is common enough that Python provides a special syntax for it called decorators.
  A decorator is a function that accepts a function as an argument and returns a new function. The new function is usually a “wrapped” version of the original function. The decorator syntax is to place an @ symbol followed by the name of the decorator function on the line before the function definition. 

```python
def compose(func1, func2): # func1 and func2 are functions
    # return a *new* function that calls func1 with the output of func2
    return lambda x: func1(func2(x))

comp = compose(lambda x: x+1, lambda x: x**2)
test = comp(5) # 5**2 + 1 = 26
print(f'result1 = {test}')
```

```python
# the same function using def rether than lambda
# define them as concepts
def outer_fun(innor_fun): # puter function is x + 1
    def warraped(x): # the argument will be passed to the innor function
        return innor_fun(x) + 1  # appling the outer function into the result of the innor function
    
    return warraped

def innor_func(x):
    return x**2

comp = outer_fun(innor_func)
test = comp(5)
print(f'result2 = {test}')
```

```python
# Doing the same using Decorator
def outer_fun(innor_fun):
    def warraped(x): # the argument will be passed to the innor function
        return innor_fun(x) + 1  # appling the outer function into the result of the innor function
    
    return warraped

@outer_fun # that is the same as:   fun2 = outer_fun(fun2)    =>  fun2(x) = outer_func(fun2(x))
def fun2(x):
    return x**2

test = fun2(5) # 5**2 + 1 = 26
print(f'result3 = {test}')
```
