##### CSV Files

```python
import csv # better to deal wiht csv files
import json

with open("week1/datasets/mpg.csv") as file:
    map = list(csv.DictReader(file)) # convert the csv to a list of dictionaries, each column as a key, each row will be an element in the list
map[:2]
```

```python
# to perform mathematicl operations, you should convert the value to float or int
# calculating the average of hwy
average = sum( [float(dic["hwy"]) for dic in map])  / len(map)
average
```

```python
# to find unique values, use set function
cylenders = set([dic["cyl"] for dic in map])
cylenders
```

```python
# we want to know the average cty depending on cylenders
cty_sum_depending_on_cly = {}
averages = {}

# iterate over the original list of dictionaries
for cyl in cylenders:

    # optaining the items which their cyl equale to the current cyl, then store their "cty" values
    cty_sum_depending_on_cly[cyl] = [ dic["cty"] for dic in list(filter(lambda dic: dic["cyl"] == cyl, map)) ]

    # get the average of each type of cyl, by sum its values and divide them by their count
    averages[cyl] = sum([float(val) for val in cty_sum_depending_on_cly[cyl]]) / len(cty_sum_depending_on_cly[cyl])
averages
```
