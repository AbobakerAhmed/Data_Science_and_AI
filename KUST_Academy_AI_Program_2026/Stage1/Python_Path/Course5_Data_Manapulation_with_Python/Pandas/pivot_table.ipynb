### Pivot Tables

the same as pivot tables in excel

df.pivot_table(
    values= 'columnName'/[list of columns],
    index= "indexColumn"/[list of index columns],
    columns= 'newColumns'/[lsit of new columns],
    aggfunc= np.stMethod/[list of stMethods]
)
return a df where
    columns are 'newColumns'/[lsit of new columns]
    index is "indexColumn"/[list of index columns]
    values= applying aggfunc/list_of_fun on 'columnName'/[list of columns] of all old rows grouped by index and columns, if list of functions, the result of each one in a single column


df.pivot_table(
    .. same ..,
    margins = True
)
return a df with the same proparities above with adding margine between columns


after making a pivot table:
pvt.index: returns all indexes of the table, if one level, a single list, in more, list of tuples

pvt.columns: returns list of tuples each with (indexVal, colmnName)



accessing:
pvt[aggfun]: return a df with all rows and columns with aggfun
pvt[aggfun][columnName]: return a series with all rows in the col

pvt[aggfun][columnName][indexVal]: a specific val
pvt[aggfun][columnName].idxmax(): max val in that series

pvt.stack(): return a df, where index as it, columns titles become a single column , each agg in a column, so pvt[index][colVal][agg]
pvt.unstack(): return it as it was: pvt[agg][colVal][index]
pvt.unstack().unstack(): return a df as: pvt[agg][colVal][index]
